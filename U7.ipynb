```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):                            
    return np.exp(-x**2) * np.sin(y)        #funktion

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)

X, Y = np.meshgrid(x, y)

Z = f(X, Y)                                 #auswertung von x,y durch funktion

plt.pcolormesh(X, Y, Z)                     #output

plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
